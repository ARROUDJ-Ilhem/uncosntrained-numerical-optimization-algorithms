#  1. Descente de Gradient à Pas Fixe
Définition de la fonction et calcul du gradient :

```python
import numpy as np

def fonction_f(x, y, z, beta):
    return x**2 + beta*x*y + y**2 + beta*y*z + z**2

def gradient_f(x, y, z, beta):
    grad_x = 2*x + beta*y
    grad_y = 2*y + beta*(x + z)
    grad_z = 2*z + beta*y
    return np.array([grad_x, grad_y, grad_z])

def descente_gradient_pas_fixe(beta, alpha=0.1, max_iter=100000, tol=1e-6):
    # Point initial
    x = 1
    y = 2
    z = 3
    
    #np.random.rand(3)  # Initialisation aléatoire
    for i in range(max_iter):
        grad = gradient_f(x, y, z, beta)
        new_x = x - alpha * grad[0]
        new_y = y - alpha * grad[1]
        new_z = z - alpha * grad[2]
        if np.linalg.norm([new_x - x, new_y - y, new_z - z]) < tol:
            break
        x, y, z = new_x, new_y, new_z
    return x, y, z

# Cas (a) : beta = 1.41
resultat_a = descente_gradient_pas_fixe(1.41)
print("Résultat avec pas fixe, beta = 1.41 :", resultat_a)

# Cas (b) : beta = 0.11
resultat_b = descente_gradient_pas_fixe(0.11)
print("Résultat avec pas fixe, beta = 0.11 :", resultat_b)
```

# 2. Descente de Gradient à Pas Variable (Recherche Linéaire)
Effectuer une recherche linéaire pour trouver un pas optimal à chaque itération.

```python
def recherche_lineaire(x, y, z, beta, grad, alpha=0.1, decay=0.8, tol=1e-6):
    # Recherche linéaire par retour en arrière
    while fonction_f(x - alpha * grad[0], y - alpha * grad[1], z - alpha * grad[2], beta) > \
          fonction_f(x, y, z, beta) - tol * alpha * np.dot(grad, grad):
        alpha *= decay
    return alpha

def descente_gradient_pas_variable(beta, max_iter=100000, tol=1e-6):
    x = 1
    y = 2
    z = 3
    for i in range(max_iter):
        grad = gradient_f(x, y, z, beta)
        alpha = recherche_lineaire(x, y, z, beta, grad)  # Pas variable
        new_x = x - alpha * grad[0]
        new_y = y - alpha * grad[1]
        new_z = z - alpha * grad[2]
        if np.linalg.norm([new_x - x, new_y - y, new_z - z]) < tol:
            break
        x, y, z = new_x, new_y, new_z
    return x, y, z

# Cas (a) : beta = 1.41
resultat_a_var = descente_gradient_pas_variable(1.41)
print("Résultat avec pas variable, beta = 1.41 :", resultat_a_var)

# Cas (b) : beta = 0.11
resultat_b_var = descente_gradient_pas_variable(0.11)
print("Résultat avec pas variable, beta = 0.11 :", resultat_b_var)

```

# 3. Méthode de Newton

```python
def methode_newton(beta, max_iter=100, tol=1e-6):
    # Initialisation
    x, y, z = np.random.rand(3)
    for i in range(max_iter):
        grad = gradient_f(x, y, z, beta)
        # Calcul de la matrice Hessienne de f
        hessian = np.array([
            [2, beta, 0],
            [beta, 2, beta],
            [0, beta, 2]
        ])
        try:
            hess_inv = np.linalg.inv(hessian)
        except np.linalg.LinAlgError:
            break  # Échec de l'inversion
        new_x, new_y, new_z = np.array([x, y, z]) - hess_inv.dot(grad)
        if np.linalg.norm(np.array([new_x - x, new_y - y, new_z - z])) < tol:
            break
        x, y, z = new_x, new_y, new_z
    return x, y, z, i

resultat_newton = methode_newton(beta=1.41)
print("Résultat méthode de Newton pour β = 1.41 :", resultat_newton)
```

# 4. Méthode quasi-Newton 

```python
from scipy.optimize import minimize

def objective_function(vars, beta):
    x, y, z = vars
    return x**2 + beta*x*y + y**2 + beta*y*z + z**2

result_quasi_newton = minimize(objective_function, [0, 0, 0], args=(1.41,), method='BFGS')
print("Résultat méthode quasi-Newton (BFGS) pour β = 1.41 :", result_quasi_newton.x)
```
